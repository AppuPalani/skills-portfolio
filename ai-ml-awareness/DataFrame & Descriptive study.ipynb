# Finding Mean median mode for T-RUNS

```python
import statistics
 
data = ipl_auction_df['T-RUNS']
 
# Finding Mean
print("\nMean: ", statistics.mean(data))
 
# Finding Median
print("Median: ", statistics.median(data))
 
# Finding Single Mode
print("Single Mode: ", statistics.mode(data))
 
# Finding Multiple Modes
print("Mode: ", statistics.multimode(data))
```

# calculating dispersion 
pstdev()

Population standard deviation of data.

pvariance()

Population variance of data.

stdev()

Sample standard deviation of data.

variance()

Sample variance of data.

```python
print("\npopulation standard deviation of data: ", statistics.pstdev(data))
print("\npopulation variance of data", statistics.pvariance(data))
print("\nsample standard deviation of data: ", statistics.stdev(data))
print("\nsample variance of data", statistics.variance(data))
```

# calulating skewness and kurtosis


```python
# as per ground thoery (various researches), cut off values are generaly accepted as -7 and +7 and skewness -3 and +3
# Importing library
from scipy.stats import skew
from scipy.stats import kurtosis
print(skew(data, axis=0, bias=False)) # Calculations are corrected for statistical bias, if 'bias' set to False. for sample
# If False, then the calculations are corrected for statistical bias.
print(kurtosis(data, axis=0, bias=False)) # Calculations are corrected for statistical bias, if 'bias' set to False. for sample
# If False, then the calculations are corrected for statistical bias.
```
